```python
df_country = pd.read_excel('Zomatodataset/Country-Code.xlsx')
```

```python
df_country
```

```python
final_df = df.merge(df_country, on='Country Code', how = 'left')
```

```python
final_df.head()
```

```python
final_df.duplicated().sum()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
# Pie Chart -- Top 3 countries that uses zomato
plt.pie(country_values[:3],labels = country_names[:3],autopct="%.2f%%",radius=2)
```

## Observation: 
### Maximum records of Zomato dataset are from India,United States,United Kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation

1. When Ratings are between 4.5 to 4.9 ---> Excellent
2. When Ratings are between 4.0 to 4.4 ---> Very Good
3. When Ratings are between 3.5 to 3.9 ---> Good
4. When Ratings are between 2.5 to 3.4 ---> Average
5. When Ratings are between 1.8 to 2.4 ---> Poor
6. When Rating is zero ---> Not Rated

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating Count' ,hue = 'Rating color', data = ratings,palette = ['blue','red','orange','yellow','green','green'])
```

## Observation
1. Not Rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x = 'Rating color',data = ratings,palette = ['blue','red','orange','yellow','green','green'])
```

```python
final_df[final_df['Aggregate rating'] == 0].groupby('Country').size().reset_index()
```
